```python
# importer le module de temps
import time

# temps de saisie en secondes
t = input("Entrez le temps en secondes: ")

# définit le compte à rebours func.
def countdown(t):
    while t:
        mins, secs = divmod(t, 60)
        timer = '{:02d}:{:02d}'.format(mins, secs)
        print(timer, end="\r")
        time.sleep(1)
        t -= 1
        print("Le temps restant est : ")
        print(t)

        print('Feu dans le trou !!')




# appel de fonction
countdown(int(t))
```
